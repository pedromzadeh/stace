```python
from stace.calibrate import ThreePoint
import matplotlib.pyplot as plt
import numpy as np

three_point_calibrator = ThreePoint([(1.3, 1), (7.6, 7), (10.8, 11)])

X = np.linspace(0, 12, 500)
X_corr = three_point_calibrator.correct(X)

plt.plot(range(500), X, label="Raw", color="orange", lw=3)
plt.plot(range(500), X_corr, label="3-point", color="green", lw=3)
plt.legend()
```
